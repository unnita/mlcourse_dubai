**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
status = data['marital-status']
uni_status = status.drop_duplicates()
uni_status
```

```python
match = {'marital-status':['Never-married','Married-civ-spouse','Divorced','Married-spouse-absent','Separated','Married-AF-spouse','Widowed'],'status':['Single','Married','Single','Married','Single','Married','Single']}
df = pd.DataFrame(match)
new_data = data.merge(df, on='marital-status', how='left')
new_data.groupby(['salary'])['status'].value_counts()
```
